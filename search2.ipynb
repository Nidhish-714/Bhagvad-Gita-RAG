```python
from tqdm.auto import tqdm
batch_size = 200

for i in tqdm(range(0, len(data), batch_size)):
    # find end of batch
    i_end = min(i+batch_size, len(data))
    # extract metadata batch
    meta_batch = data.iloc[i:i_end]
    meta_dict = meta_batch.to_dict(orient="records")
    # concatinate all metadata field except for id and year to form a single string
    meta_batch = [" ".join(x) for x in meta_batch.loc[:, ~meta_batch.columns.isin(['id', 'year'])].values.tolist()]
    # extract image batch
    img_batch = data[i:i_end]
    # create sparse BM25 vectors
    sparse_embeds = bm25.encode_documents([text for text in meta_batch])
    # create dense vectors
    dense_embeds = model.encode(img_batch).tolist()
    # create unique IDs
    ids = [str(x) for x in range(i, i_end)]

    upserts = []
    # loop through the data and create dictionaries for uploading documents to pinecone index
    for _id, sparse, dense, meta in zip(ids, sparse_embeds, dense_embeds, meta_dict):
        upserts.append({
            'id': _id,
            'sparse_values': sparse,
            'values': dense,
            'metadata': meta
        })
    # upload the documents to the new hybrid index
    index.upsert(upserts)

# show index description after uploading the documents
index.describe_index_stats()
```

```python
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize
import nltk

# Ensure necessary NLTK data is downloaded
nltk.download('punkt')

class HybridSearch:
    # Existing initialization code remains the same

    def __init__(self, pinecone_api_key: str, pinecone_host: str, mongo_uri: str,
                 db_name: str, collection_name: str):
        # Existing initialization code remains unchanged
        # New BM25 initialization
        self.bm25_corpus = []
        self.bm25_docs = []
        self._prepare_bm25()

    def _prepare_bm25(self):
        """Prepare BM25 model for sparse search."""
        cursor = self.collection.find({}, {"_id": 1, "fullplot": 1, "translation_in_english": 1, "meaning_in_hindi": 1})
        for doc in cursor:
            fields = [doc.get("fullplot", ""), doc.get("translation_in_english", ""), doc.get("meaning_in_hindi", "")]
            text = " ".join(fields)
            tokenized_text = word_tokenize(text.lower())
            self.bm25_corpus.append(tokenized_text)
            self.bm25_docs.append(doc)
        
        self.bm25_model = BM25Okapi(self.bm25_corpus)

    def sparse_search(self, query: str, top_k: int = 10) :
        """Perform sparse search using BM25."""
        tokenized_query = word_tokenize(query.lower())
        scores = self.bm25_model.get_scores(tokenized_query)
        
        ranked_indices = np.argsort(scores)[::-1][:top_k]
        sparse_results = []
        
        for idx in ranked_indices:
            if scores[idx] > 0:  # Include only documents with non-zero scores
                doc = self.bm25_docs[idx]
                doc["sparse_score"] = scores[idx]
                sparse_results.append(doc)
        
        return sparse_results

    def hybrid_search(self, query: str, w_semantic: float = 0.5, w_keyword: float = 0.5, top_k: int = 10) :
        """Combine semantic search with sparse search."""
        semantic_results = self.semantic_search(query, top_k=top_k)
        sparse_results = self.sparse_search(query, top_k=top_k)

        all_docs = {}

        # Combine semantic results
        for doc in semantic_results:
            doc_id = str(doc['_id'])
            if doc_id not in all_docs:
                all_docs[doc_id] = {
                    'doc': doc,
                    'semantic_score': doc['semantic_score'],
                    'sparse_score': 0
                }

        # Combine sparse results
        for doc in sparse_results:
            doc_id = str(doc['_id'])
            if doc_id in all_docs:
                all_docs[doc_id]['sparse_score'] = doc['sparse_score']
            else:
                all_docs[doc_id] = {
                    'doc': doc,
                    'semantic_score': 0,
                    'sparse_score': doc['sparse_score']
                }

        # Calculate hybrid scores
        results = []
        for doc_id, info in all_docs.items():
            hybrid_score = (w_semantic * info['semantic_score'] +
                            w_keyword * info['sparse_score'])
            results.append({
                'doc': info['doc'],
                'hybrid_score': hybrid_score,
                'semantic_score': info['semantic_score'],
                'sparse_score': info['sparse_score']
            })

        results.sort(key=lambda x: x['hybrid_score'], reverse=True)
        return results[:top_k]
```
